```python
import easyocr
reader = easyocr.Reader(['en'])
```

```python
import PIL
from PIL import ImageDraw
im = PIL.Image.open("/content/House.jpeg")
im
```

```python
bounds = reader.readtext("/content/House.jpeg")
bounds
```

```python
def draw_boxes(image, bounds, color='yellow', width=2):
  draw = ImageDraw.Draw(image)
  for bound in bounds:
    p0, p1, p2, p3 = bound[0]
    draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
  return image

draw_boxes(im, bounds)
```

```python
bounds = reader.readtext('/content/House.jpeg', contrast_ths=0.05, adjust_contrast=0.7, add_margin=0.45, width_ths=0.7, decoder='beamsearch')
bounds
```
